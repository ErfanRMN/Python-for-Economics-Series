# <font face="gotham" color="navy">1. Line Plot

```python
x1 = np.linspace(-4 * np.pi, 4 * np.pi, 1000)
y1 = np.cos(x1)

plt.figure(figsize=(16,7))
plt.plot(x1, y1, color='blue', label='cos(x)', lw=1.5)
plt.grid(ls = ':', alpha = 0.25)

plt.axhline(0, lw=2, ls="--", c='black')
plt.axvline(0, lw=2, ls="--", c='black')

plt.xlim((-3*np.pi,2*np.pi))
# plt.ylim()

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('$ cos(x) $ plot within (-3, 2) interval', fontsize=24)
plt.xlabel('$ x $', fontsize=20)
plt.ylabel('$ cos(x) $', fontsize=20)

plt.legend(fontsize=18,frameon=True)

#plt.savefig("cosxplot.svg", dpi=1000)

plt.show()
```

# <font face="gotham" color="navy">2. Bar Plot (vertical)

```python
cat2 = ['A', 'B', 'C', 'D']
val2 = [15, 30, 60, 10]
col2 = ['blue', 'green', 'red', 'orange']

plt.figure(figsize=(16,7))
plt.bar(cat2, val2, color=col2)
plt.grid(ls=":",alpha=0.25)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Bar Plot (vertical)',fontsize=24)
plt.xlabel('Categories',fontsize=20)
plt.ylabel('Values',fontsize=20)

plt.show()
```

# <font face="gotham" color="navy">3. Bar Plot (horizontal)

```python
categories3 = ['A', 'B', 'C', 'D']
values3 = [15, 30, 60, 10]
colors3 = ['blue', 'green', 'red', 'orange']

plt.figure(figsize=(16,7))
plt.barh(categories3, values3, color=colors3)
plt.grid(ls=":",alpha=0.25)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Bar Plot (horizontal)',fontsize=24)
plt.xlabel('Values',fontsize=20)
plt.ylabel('Categories',fontsize=20)

plt.show()
```

# <font face="gotham" color="navy">4. Box Plot (vertical)

```python
d4 = np.random.normal(0, 1, 100)

plt.figure(figsize=(17,7))

box4 = {'color': 'red', 'linestyle': ':', 'linewidth': 2}
whi4 = {'color': 'violet', 'linestyle': '-.', 'linewidth': 1.5}
cap4 = {'color': 'blue', 'linestyle': ':', 'linewidth': 1}
fli4 = {'color': 'purple', 'marker': '2', 'markersize': 5}
med4 = {'color': 'green', 'linestyle': '-', 'linewidth': 2}

plt.boxplot(d4, medianprops=med4, flierprops=fli4, capprops=cap4,
           whiskerprops=whi4, boxprops=box4)

plt.xticks([])
plt.ylabel('Data', fontsize=18)
plt.title('Box Plot', fontsize=18)

plt.show()
```

# <font face="gotham" color="navy">5. Box Plot (horizontal)

```python
d5 = np.random.normal(0, 1, 100)

plt.figure(figsize=(17,7))

box5 = {'color': 'red', 'linestyle': '--', 'linewidth': 2}
whi5 = {'color': 'violet', 'linestyle': '-.', 'linewidth': 1.5}
cap5 = {'color': 'blue', 'linestyle': ':', 'linewidth': 1}
fli5 = {'color': 'purple', 'marker': 'x', 'markersize': 5}
med5 = {'color': 'green', 'linestyle': '-', 'linewidth': 2}

plt.boxplot(d5, medianprops=med5, flierprops=fli5, capprops=cap5,
           whiskerprops=whi5, boxprops=box5, vert=False)

plt.yticks([])
plt.xlabel('Data', fontsize=18)
plt.title('Box Plot', fontsize=18)

plt.show()
```

# <font face="gotham" color="navy">6. Pie Plot

```python
lab6 = ['A', 'B', 'C', 'D']
siz6 = [15, 30, 45, 10]
col6 = ['yellow', 'green', 'cyan', 'red']
exp6 = (0.1, 0.1, 0.1, 0.1)
font6 = {'fontsize': 18}

plt.figure(figsize=(17,8))

plt.pie(siz6, explode=exp6, labels=lab6, colors=col6,
       autopct='%1.1f%%', shadow=True, startangle=60, textprops=font6)

plt.axis('equal')

plt.title('Pie Chart', fontsize=24)

plt.show()
```

# <font face="gotham" color="navy">7. Donut Plot

```python
lab7 = ['A', 'B', 'C', 'D']
siz7 = [15, 30, 45, 10]
col7 = ['yellow', 'lime', 'cyan', 'red']
radius7 = 0.6
font7 = {'fontsize': 18}

plt.figure(figsize=(16,7))


plt.pie(siz7, labels=lab7, colors=col7, autopct='%1.1f%%', textprops=font7,
       startangle=90, wedgeprops=dict(width=radius7), shadow=True)

plt.axis('equal')

plt.title('Donut Plot', fontsize=24)

plt.show()
```

# <font face="gotham" color="navy">8. Scatter Plot

```python
x8 = np.random.rand(100)
y8 = np.random.rand(100)

plt.figure(figsize=(16,7))

plt.scatter(x8, y8, color='b', marker='x')

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Scatter Plot', fontsize=24)
plt.xlabel('X', fontsize=20)
plt.ylabel('Y', fontsize=20)

plt.show()
```

# <font face="gotham" color="navy">9. Connected Scatter Plot

```python
x9 = np.random.rand(10)
y9 = np.random.rand(10)

fig, ax = plt.subplots(figsize=(8, 6))

ax.scatter(x9, y9, color='b', marker='D', label='Data Points')

ax.plot(x9, y9, color='r', linestyle='-', linewidth=1, label='Connected Lines')

ax.set_title('Connected Scatter Plot', fontsize = 16)
ax.set_xlabel('X', fontsize = 15)
ax.set_ylabel('Y', fontsize = 15)

ax.legend()

plt.show()
```

# <font face="gotham" color="navy">10. Dual-Axis Plot

```python
x10 = [1, 2, 3, 4, 5]
bard10 = [10, 15, 7, 12, 8]
lind10 = [12, 15, 7, 12, 8]

fig, ax = plt.subplots()

ax.bar(x10, bard10, color='blue', label='Bar Plot')
ax.set_ylabel('Primary Y-axis')

ax2 = ax.twinx()

ax2.plot(x10, lind10, color='red', label='Line Plot')
ax2.set_ylabel('Secondary Y-axis')

ax.set_xlabel('X-axis')
ax.set_title('Dual-Axis Bar-Line Plot')

ax.legend(loc='upper left', bbox_to_anchor=(0.0, 1.15))

plt.show()
```

# <font face="gotham" color="navy">11. Dendrogram

```python
from scipy.cluster.hierarchy import dendrogram, linkage

x11 = np.random.rand(10, 2)

z11 = linkage(x11, method='ward')

plt.figure(figsize=(10, 6))
dendrogram(z11)

plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')

plt.show()
```

# <font face="gotham" color="navy">12. Tree Map

```python
import squarify

d12 = [10, 5, 3, 2, 1]

lab12 = ['A', 'B', 'C', 'D', 'E']

col12 = ['violet', 'navy', 'tomato', 'lime', 'orange']

fig, ax = plt.subplots()
squarify.plot(sizes=d12, label=lab12, color=col12, alpha=0.7)

ax.axis('off')
ax.set_title('Tree Map')

plt.show()
```

# <font face="gotham" color="navy">13. Waffle Plot

```python
from pywaffle import Waffle

d13 = {'A': 20, 'B': 20, 'C': 20, 'D': 20}
cat13 = list(d13.keys())

fig = plt.figure(
    FigureClass=Waffle,
    rows=20,
    columns=20,
    values=d13,
    labels=cat13,
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)}
)

plt.title('Waffle Plot')

plt.show()
```
